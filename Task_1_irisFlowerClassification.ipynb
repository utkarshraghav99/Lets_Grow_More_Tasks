```python
x=df.drop('class',axis=1)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
logistic_grid={"C":np.logspace(-3,3,7),"penalty":["l1","l2","none"]}
logistic=LogisticRegression()
logistic_CV=GridSearchCV(logistic,logistic_grid,cv=10)
logistic_CV.fit(x_train,y_train)
```

```python
print("Hyperparameters : (best parameters)", logistic_CV.best_params_)
print("Best Accuracy:",logistic_CV.best_score_)
```

```python
logistic_best=LogisticRegression(C=1.0,penalty='l2')
logistic_best.fit(x_train,y_train)
logistic_predict=logistic_best.predict(x_test)
```

```python
print('Accuracy: ',accuracy_score(y_test,logistic_predict))
```

```python
confusion= confusion_matrix(y_test,logistic_predict)
sns.heatmap(confusion, annot=True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
random_grid = {'criterion' :['gini', 'entropy'],
               'n_estimators':[50,150],
               'max_depth': np.arange(10,20),
               'bootstrap': [True, False]}
random = RandomForestClassifier()
random_cv=GridSearchCV(random,random_grid,cv=None)
random_cv.fit(x_train,y_train)
```

```python
print("Hpyerparameters :(best parameters) ",random_cv.best_params_)
print("Best Accuracy :",random_cv.best_score_)
```

```python
random_best= RandomForestClassifier(bootstrap=True, criterion='gini', max_depth=10,n_estimators=50)
random_best.fit(x_train, y_train)
random_predict=random_best.predict(x_test)
```

```python
print('Accuracy score:',accuracy_score(y_test, random_predict))
```

```python
confusion = confusion_matrix(y_test,random_predict)
sns.heatmap(confusion, annot=True)
plt.show()
```

```python
from sklearn.svm import SVC

svc_grid={'C':[1,10,100,1000],
          'gamma':['scale','auto'],
          'kernel':['linear','rbf']}

svc = SVC()
svc_cv=GridSearchCV(svc,svc_grid,cv=None)
svc_cv.fit(x_train,y_train)
```

```python
print("Tuned hpyerparameters :(best parameters) ",svc_cv.best_params_)
print("Best Accuracy :",svc_cv.best_score_)
```

```python
svc_best= SVC(C=1, gamma='scale', kernel='linear')
svc_best.fit(x_train, y_train)
svc_predict=svc_best.predict(x_test)
```

```python
print('Accuracy score:',accuracy_score(y_test, svc_predict))
```

```python
confusion = confusion_matrix(y_test,svc_predict)
sns.heatmap(confusion, annot=True)
plt.show()
```
